# Decorator Endpoint

This endpoint takes in a knowledge graph from another reasoner, and augments its edges it with sentences from GNBR.

```python
import json
import requests

def pubs(question):
    url=f'https://gnbr-reason.ncats.io/decorator'
    print(url)
    response = requests.post(url,json=question)
    print( f"Return Status: {response.status_code}" )
    if response.status_code == 200:
        return response.json()
    return {}
```

##### Load WF1 module 1 output from rtx

Import data and do some massaging to bring up to 0.9 standard

```python
with open('wf1mod1_1_ebola_gamma.txt') as f:
    rtx = json.load(f)

# Specify result number in range 0-4
index = 0
    
rtx_result = rtx['result_list'][index]

for edge in rtx_result['result_graph']['edge_list']:
    edge['publications'] = list(edge['publications'].split(','))

for node in rtx_result['result_graph']['node_list']:
    node['type'] = [node['type']]
    
node_list = rtx_result['result_graph']['node_list']
edge_list = rtx_result['result_graph']['edge_list']

query_graph = {
    'nodes': node_list, 
    'edges': edge_list
}
```

##### Construct query object

```python
query = {
    "query_message": {
        "query_graph": query_graph
    }
}
```

##### Post Query to Endpoint

```python
KG = pubs(query)['results'][0]['result_graph']
```
