```python
#### Import some needed modules
import requests
import json
import sys
import pandas as pd
```

```python
robokop_server = 'robokop.renci.org'
def synonymize(nodetype,identifier):
    url=f'http://{robokop_server}:6010/api/synonymize/{identifier}/{nodetype}/'
    response = requests.post(url)
    if response.status_code == 200:
        return response.json()
    return []
```

```python
#### Set the base URL for the reasoner and its endpoint
API_BASE_URL = 'https://rtx.ncats.io/api/rtx/v1'
url_str = API_BASE_URL + "/query"
```

```python
#### Create a dict of the request, specifying the the query type to Q50 - "Which biological processes does $protein participate in"
#### Set protein to the curie id of the desired protein and run.
request = { "query_type_id": "Q50", "terms": { "protein": protein } }
```

```python
#### Send the request to RTX and check the status
response_content = requests.post(url_str, json=request, headers={'accept': 'application/json'})
status_code = response_content.status_code
assert status_code == 200
```

```python
#### Unpack the response content into a dict
response_dict = response_content.json()
```

```python
#BP
goTermsBP=list()
if not(response_dict['n_results']==0):
    for result in response_dict["result_list"]:
        goTermsBP.append(result["row_data"][3])
```

```python
#### Create a dict of the request, specifying the the query type to Q56 - "Which molecular function is $protein capable of"
#### Set protein to the curie id of the desired protein and run.
request = { "query_type_id": "Q56", "terms": { "protein": protein } }
```

```python
#### Send the request to RTX and check the status
response_content = requests.post(url_str, json=request, headers={'accept': 'application/json'})
status_code = response_content.status_code
assert status_code == 200
```

```python
#### Unpack the response content into a dict
response_dict = response_content.json()
```

```python
goTermsMF=list()
if not(response_dict['n_results']==0):
    for result in response_dict["result_list"]:
        goTermsMF.append(result["row_data"][3])
    print(goTermsMF)
```

```python
bpRequests=list()
bpResponses=list()
for BP in goTermsBP:
    request = { "query_type_id": "Q51", "terms": { "biological_process": BP } }
    bpRequests.append(request)
```

```python
for request in bpRequests:
    response_content = requests.post(url_str, json=request, headers={'accept': 'application/json'})
    status_code = response_content.status_code
    assert status_code == 200
    response_dict = response_content.json()
    bpResponses.append(response_dict)
    if "table_column_names" in response_dict:
      print("\t".join(response_dict["table_column_names"]))
      for result in response_dict["result_list"]:
        print("\t".join(result["row_data"]))
```

```python
mfRequests=list()
mfResponses=list()
for MF in goTermsMF:
    request = { "query_type_id": "Q58", "terms": { "molecular_function": MF } }
    mfRequests.append(request)
```

```python
for request in mfRequests:
    response_content = requests.post(url_str, json=request, headers={'accept': 'application/json'})
    status_code = response_content.status_code
    assert status_code == 200
    response_dict = response_content.json()
    mfResponses.append(response_dict)
    if "table_column_names" in response_dict:
      print("\t".join(response_dict["table_column_names"]))
      for result in response_dict["result_list"]:
        print("\t".join(result["row_data"]))
```

```python
answers = {}
for response in bpResponses:
    if "table_column_names" in response:
        for result in response["result_list"]:
            sourceName =result["row_data"][0]
            sourceId = result["row_data"][1]
            targetName=result["row_data"][2]
            targetId=result["row_data"][3]
            synonyms = synonymize('gene',result["row_data"][3])
            answers.setdefault(synonyms["name"],{})[sourceName]=1
            #print(synonyms["name"]+" ("+synonyms["id"]+") similar by way of "+sourceName)
```
