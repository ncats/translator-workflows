```python
handle = Entrez.einfo(db="gene")
record = Entrez.read(handle)
print(record["DbInfo"]["Description"])
print(record["DbInfo"]["Count"])
print(record["DbInfo"]["LastUpdate"])
```

```python
for field in record["DbInfo"]["FieldList"]:
    print("%(Name)s, %(FullName)s, %(Description)s" % field)
```

```python
handle = Entrez.esearch(db="gene", term="asthma")
record = Entrez.read(handle)
idList = record["IdList"]
print(idList)
```

```python
for entry in idList:
    handle = Entrez.efetch( db="gene", id=entry, rettype="gene_table", retmode="text" )
    print(handle.read())
    break
```
