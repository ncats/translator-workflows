# Workflow 1, Module 2, Question 1 and 2

## What genes are implicated in [condition]?
## What subset of genes are most representative of [conditions]?  (find archetypes)

ROBOKOP can do these two questions together as an enriched expansion from condition to gene.

```python
robokop_server = 'robokop.renci.org'
```

```python
import requests
import json
import pandas as pd
```

```python
def enrichment(type1,identlist,type2,threshhold=None,maxresults=None,numtype1=None,include_descendants=None,rebuild=None):
    url=f'http://{robokop_server}/api/simple/enriched/{type1}/{type2}'
    params = { 'threshhold': threshhold, 'maxresults': maxresults, 
              'num_type1':numtype1, 'identifiers': identlist, 
              'include_descendants':include_descendants, 'rebuild': rebuild }
    params = { k:v for k,v in params.items() if v is not None }
    response=requests.post(url, json = params)
    print( f'Return Status: {response.status_code}' )
    if response.status_code == 200:
        return response.json()
    return []
```

These are some of the candidate conditions from module 1

```python
glucose_intolerance='MONDO:0001076'
mody='MONDO:0018911'
cystic_fibrosis='MONDO:0009061'
pcd='MONDO:0009484'
```

```python
glucose_intolerance_genes = enrichment('disease',[glucose_intolerance],'gene',include_descendants=True)
pd.DataFrame(glucose_intolerance_genes)
```

```python
mody_genes = enrichment('disease',[mody],'gene',include_descendants=True)
pd.DataFrame(mody_genes)
```

```python
cystic_fibrosis_genes = enrichment('disease',[cystic_fibrosis],'gene',include_descendants=True)
pd.DataFrame(cystic_fibrosis_genes)
```
