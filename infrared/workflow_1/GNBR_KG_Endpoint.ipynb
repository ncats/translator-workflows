# GNBR Knowledge Graph Endpoint

This endpoint takes in a knowledge graph from another reasoner, and returns a network supporting setnences from GNBR.

```python
import json
import requests

def post(question):
    url=f'https://gnbr-reason.ncats.io/gnbr'
    print(url)
    response = requests.post(url,json=question)
    print( f"Return Status: {response.status_code}" )
    if response.status_code == 200:
        return response.json()
    return {}
```

##### Load WF1 module 1 output from rtx

```python
with open('wf1mod1_1_ebola_gamma.txt') as f:
    rtx = json.load(f)

for edge in rtx['result_list'][0]['result_graph']['edge_list']:
    edge['publications'] = list(edge['publications'].split(','))

node_list = rtx['result_list'][0]['result_graph']['node_list']
edge_list = rtx['result_list'][0]['result_graph']['edge_list']

query_graph = {
    'nodes': node_list, 
    'edges': edge_list
}
```

##### Construct query object

```python
query = {
    "query_message": {
        "query_graph": query_graph
    }
}
```

##### Post Query to Endpoint

```python
result_graph = post(query)['results'][0]['result_graph']
```
